### Obtención del atributo edad
Debido a que se quiere realizar un análisis usando la edad de los estudiantes, y solamente se tiene la fecha de nacimiento se inicia un proceso de trabajo con dicho atributo

````python
df_clean['ESTU_PRUEBA'] = df_clean['ESTU_FECHANACIMIENTO'].str.extract('([0-9]{1,2}/[0-9]{1,2}/....)')
````

````python
df_clean['ESTU_FECHA'] = pd.to_datetime(df_clean['ESTU_PRUEBA'],errors = 'coerce')
````

````python
df_clean[df_clean.isin([np.nan, np.inf, -np.inf]).any(1)]
````

````python
from datetime import date
def extract_age(born):
    app_date = date(2019, 10, 20)
    return app_date.year - born.year - ((app_date.month, app_date.day) < (born.month, born.day))
    
````

````python
df_clean['ESTU_EDAD'] = df_clean['ESTU_FECHA'].map(lambda born_date: extract_age(born_date))
````

````python
df_clean = df_clean.drop(['ESTU_FECHANACIMIENTO','ESTU_COD_RESIDE_DEPTO', 'INST_COD_INSTITUCION', 'ESTU_SNIES_PRGMACADEMICO', 'ESTU_PRUEBA'], axis=1).copy()
````

Para iniciar el análisis de dos variables, dado que se tiene en su mayoría variables categóricas y al número de variables que se tienen, se van a relacionar solo ciertas variables de interés. Una forma de analizar dos categorías en forma conjunta es usando una tabla de doble entrada, o tabla de contingencia.

````python
#Pruebas para trabajar análisis de dos categorías
pd.crosstab(index = df_clean['ESTU_DEPTO_RESIDE'],
           columns = df_clean['ESTU_GENERO'],
           margins = True)
````

Para el ejemplo se va a analizar la relación entre la edad y el departamento de residencia. Los márgenes derecho e inferior de la tabla muestran los totales. El registro inferior de la tabla representa la distribución de frecuencia del género de los estudiantes. La última columna es la distribución de frecuencia de la variable departamentos. Cada celda de cada márgen representa la distribución marginal. Cada celda es el recuento para la combinación de los valores de las dos variables categóricas

````python
pd.crosstab(index = df_clean['ESTU_DEPTO_RESIDE'],
           columns = df_clean['ESTU_GENERO'],
           margins = True).apply(lambda r: r/len(df_clean) * 100,
                                axis = 1)
````

````python
pd.crosstab(index = df_clean['ESTU_DEPTO_RESIDE'],
           columns = df_clean['ESTU_GENERO']).apply(lambda r: r/r.sum() * 100,
                                                  axis = 1)
````

De aquí se puede ver que de los estudiantes que son de Bogotá, el 57.8% son mujeres y el 42.48% son hombres.

````python
pd.crosstab(index = df_clean['ESTU_DEPTO_RESIDE'],
           columns = df_clean['ESTU_GENERO']).apply(lambda r: r/r.sum() * 100,
                                                  axis = 0)
````

Ésta última tabla de porcentajes relativos son en relación al total de cada género y así se puede establecer que del total de estudiantes mujeres que presentaron la prueba el 27.4% es de residencia de Bogotá y del total de hombres que realizaron la prueba, el 29.2 son de residencia en Bogotá.

## Análisis predictivo

Como introducción al análisis predictivo se quiere realizar un árbol de clasificación utilizando las características económicas del estudiante, para lo cual se utilizan doce atributos que describen este aspecto

````python
df_clean['PUNT_RANGO'] = pd.qcut(df_clean['PUNT_GLOBAL'], q = 4)
````

````python
df_clean['PUNT_RANGO'].value_counts().plot(kind='bar')
````

````python
df_clean.info()
````

````python
df_toPredict = df_clean[['ESTU_AREARESIDE', 'ESTU_VALORMATRICULAUNIVERSIDAD','ESTU_PAGOMATRICULABECA','ESTU_PAGOMATRICULACREDITO', 'ESTU_PAGOMATRICULAPADRES','ESTU_PAGOMATRICULAPROPIO','FAMI_ESTRATOVIVIENDA','FAMI_TIENEINTERNET','FAMI_TIENECOMPUTADOR','FAMI_TRABAJOLABORPADRE','FAMI_TRABAJOLABORMADRE', 'PUNT_RANGO']].copy()
````

````python
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from IPython.display import Image as PImage
from subprocess import check_call
from PIL import Image, ImageDraw, ImageFont
````

Para poder ejecutar el algoritmo es necesario codificar los atributos, para lo cual se hace uso de la función labelEnconder

````python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
````

Para agilizar el proceso se codifica columna a columna del dataframe a usar para la generación del árbol

````python
for i in df_toPredict.columns:
    df_toPredict[i] = (lb.fit_transform(df_toPredict[i].astype(str)))
````

````python
df_toPredict
````

Se va a realizar un recorrido por todos los atributos, construyendo distintos árboles para mirar las diferentes exactitudes y así encontrar el árbol de mayor valor

````python
cv = KFold(n_splits=5) # Numero deseado de "folds" que haremos
accuracies = list()
max_attributes = len(list(df_toPredict))
depth_range = range(1, max_attributes + 1)

# Testearemos la profundidad de 1 a cantidad de atributos +1
for depth in depth_range:
    fold_accuracy = []
    tree_model = tree.DecisionTreeClassifier(criterion='gini',
                                             max_depth = depth)
    for train_fold, valid_fold in cv.split(df_toPredict):
        f_train = df_toPredict.iloc[train_fold] 
        f_valid = df_toPredict.iloc[valid_fold] 

        model = tree_model.fit(X = f_train.drop(['PUNT_RANGO'], axis=1), 
                               y = f_train["PUNT_RANGO"]) 
        valid_acc = model.score(X = f_valid.drop(['PUNT_RANGO'], axis=1), 
                                y = f_valid["PUNT_RANGO"]) # calculamos la precision con el segmento de validacion
        fold_accuracy.append(valid_acc)

    avg = sum(fold_accuracy)/len(fold_accuracy)
    accuracies.append(avg)
    
# Mostramos los resultados obtenidos
df = pd.DataFrame({"Max Depth": depth_range, "Average Accuracy": accuracies})
df = df[["Max Depth", "Average Accuracy"]]
print(df.to_string(index=False))
````

Del resultado se puede apreciar que el nivel que ofrece la mejor exactitud es el 9, por lo que se procede a consturir el árbol particular

````python
y_train = df_toPredict['PUNT_RANGO']
x_train = df_toPredict.drop(['PUNT_RANGO'], axis = 1).values

decision_tree = tree.DecisionTreeClassifier(criterion = 'gini',
                                            min_samples_split=20,
                                            min_samples_leaf=5,
                                            max_depth = 9,
                                            class_weight='balanced')
decision_tree.fit(x_train, y_train)
````

````python
acc_decition_tree = round(decision_tree.score(x_train, y_train) * 100, 2)
acc_decition_tree
````
