### Proceso de limpieza
Primero se evaluó el porcentaje de celdas faltantes

```python
missing_values_count = df.isnull().sum()
total_cells = np.product(df.shape)
total_missing_values = missing_values_count.sum()
missing_values_rate = total_missing_values / total_cells * 100
print(missing_values_rate)
```

```python
intervencion_si = len(df[df["Intervencion quirurgica 1) si. 2) no."] == 0])
intervencion_no = len(df[df["Intervencion quirurgica 1) si. 2) no."] == 1])
plt.pie([intervencion_si, intervencion_no], labels=["Si", "No"],autopct="%0.1f %%")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
vars_to_look = ['Fiebre alta en el ultimo anio 1) hace menos de tres meses. 2) hace mas de tres meses. 3) no.','Frecuencia de consumo de alcohol 1) varias veces al dia. 2) todos los dias. 3) varias veces a la semana. 4) una vez a la semana. 5) casi nunca o nunca']
for i, var in enumerate(vars_to_look):
    plt.subplot(1,2,i+1)
    plt.xticks(rotation = 90)
    plt.subplots_adjust(hspace=0.5)
    plt.subplots_adjust(wspace = 0.5)
    sns.distplot(df[var])
    plt.title(var, fontsize=14, verticalalignment='bottom')
```

```python
Infantiles = len(df[df["Enfermedades infantiles (varicela. sarampion. paperas. polio) 1) si. 2) no."] == 1])
Traumatismo = len(df[df["Accidente o traumatismo grave 1) si. 2) no. (0. 1)"] == 1])
Diagnostico = len(df[df["Salida: Diagnostico normal (N). alterado (O)"] == 'O'])

barWidth = 0.25

br1 = np.arange(1)
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, Infantiles, color ='r', width = barWidth,
        edgecolor ='grey', label ='Infantiles')
plt.bar(br2, Traumatismo, color ='g', width = barWidth,
        edgecolor ='grey', label ='Traumatismo')
plt.bar(br3, Diagnostico, color ='b', width = barWidth,
        edgecolor ='grey', label ='Diagnostico alterado')

plt.xlabel('Si', fontweight ='bold', fontsize = 15)
plt.ylabel('Cantidad de registros', fontweight ='bold', fontsize = 15)

plt.legend()
plt.show()
```
