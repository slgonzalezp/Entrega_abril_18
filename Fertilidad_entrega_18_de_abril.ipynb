```python
def clasificar_edad(x):
    if isinstance(x, float):
        if 0.0 <= x<=0.33:
            return 'entre_18_24'
        if 0.34 <= x <= 0.66:
            return 'entre_24_30'
        if 0.67 <= x <= 1.0:
            return 'entre_30_36'
    else:
        return x
df['Edad en el momento del analisis 18-36'] = df['Edad en el momento del analisis 18-36'].apply(clasificar_edad)
def clasificar_hora_sentado(x):
    if isinstance(x, float):
        if 0.0 <= x<=0.5:
            return 'entre_0_12'
        if 0.51 <= x <= 1.0:
            return 'entre_12_24'

    else:
        return x
df['Numero de horas sentadas al dia ene-16 '] = df['Numero de horas sentadas al dia ene-16 '].apply(clasificar_hora_sentado)
```

```python
df["Salida Diagnostico normal (N). alterado (O)"] = df["Salida Diagnostico normal (N). alterado (O)"].replace('N','Normal')
df["Salida Diagnostico normal (N). alterado (O)"] = df["Salida Diagnostico normal (N). alterado (O)"].replace('O','Alterado')
```

AdaBoost son los árboles de decisión con un nivel, es decir, con árboles de decisión con solo 1 división. Estos árboles también se llaman tocones de decisión.
Lo que hace este algoritmo es que construye un modelo y otorga pesos iguales a todos los puntos de datos. Luego asigna pesos más altos a los puntos que están mal clasificados. Ahora todos los puntos que tienen pesos más altos tienen más importancia en el próximo modelo.El árbol con el Índice Gini más bajo será nuestro primer tocón. En el gráfico se observa la escasez de errores de clasificación 0,4. Sabemos que la suma total de los pesos muestrales debe ser igual a 1, pero aquí, si sumamos todos los pesos muestrales nuevos, obtendremos 0.0372.Para que esta suma sea igual a 1, normalizaremos estos pesos dividiendo todos los pesos por la suma total de pesos actualizados que es 0372.Entonces, después de normalizar los pesos de la muestra, obtenemos este conjunto de datos.
